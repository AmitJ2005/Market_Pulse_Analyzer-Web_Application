```python
from yahooquery import search

# Replace 'COMPANY_NAME_TO_CHECK' with the specific company name you want to check
company_name_to_check = ' Tata Power Company'

try:
    # Use Yahoo Finance search to get stock symbols based on the company name
    result = search(company_name_to_check)

    if result['quotes']:
        # Take the first stock symbol from the search results
        stock_symbol = result['quotes'][0]['symbol']
        print(f"Stock symbol for {company_name_to_check}: {stock_symbol}")
    else:
        # Handle the case where no stock symbol is found
        print(f"No stock symbol found for {company_name_to_check}")
except Exception as e:
    print(f"Error fetching stock symbol for {company_name_to_check}: {e}")
```

```python
import yfinance as yf

# Specify the stock symbol and date range
stock_symbol = "TATAPOWER.BO"
start_date = "2023-01-01"
end_date = "2023-12-31"

# Fetch stock data with one-day interval
stock_data_daily = yf.download(stock_symbol, start=start_date, end=end_date, interval='1d')

# Print the stock data
stock_data_daily
```

```python
import yfinance as yf
import pandas as pd

# Replace 'RELIANCE.BSE' with the symbol you are interested in
symbol = 'RELIANCE.BO'

# Download historical data as a pandas DataFrame
data = yf.download(symbol, period='max')

# Reset the index to have the 'Date' as a column
df = data.reset_index()

# Display the DataFrame
print(df.head())
```
